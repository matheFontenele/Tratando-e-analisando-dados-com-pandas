#Aplicando filtros

Apartamentos que possuem 1 quarto e aluguel menor que R$ 1200,00

```python
selecao1 = apartamentos['Quartos'] == 1 #Retorna um valor booleano onde True caso o numero de quartos seja 1
apartamentos[selecao1]
```

```python
selecao2 = apartamentos['Valor'] < 1200 # Uandao o mesmo metodo acima para filtras os apartamentos onde o valor seja menor que 1200
apartamentos[selecao2]
```

```python
selecao_final = (selecao1) & (selecao2) #Junta as duas condições acima em uma unica variavel
apartamentos_1quarto_menorq1200 = apartamentos[selecao_final].sort_values('Valor')
```

Apartamentos que possuem pelo menos 2 quartos, aluguel menor que R$ 3000,00 e área maior que 70 m quadrados

```python
selecao = (apartamentos['Quartos'] >=2) & (apartamentos['Valor'] < 3000) & (apartamentos['Area'] > 70) # Filtro direto das condições em uma unica linha
apartamentos_2quartos_menorq3000_area70 = apartamentos[selecao]
```

#Criando colunas numericas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
```

valor_por_mes: essa coluna deve conter os gastos mensais de cada imóvel, incluindo aluguel e condomínio;

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio'] #desta forma, criamos uma nova coluna com a soma dos valores das colunas valor e condominio
dados.head()
```

valor_por_ano: essa coluna deve conter os gastos anuais por imóvel, ou seja, IPTU mais 12 meses de aluguel e condomínio.

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

#Criando uma coluna categorica

Descrição: essa coluna deve possuir uma sumarização das principais informações dos imóveis que serão apresentadas no site: tipo de imóvel, bairro, quantidade de quartos e vagas de garagem;

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quarto(s) e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
#Metodo astype converte o tipo de uma variavel
dados.head()
```

Possui_suite: essa deve ser uma coluna que informe apenas se o imóvel possui ou não suítes, sem se importar com a quantidade.

```python
dados['possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
# metodo apply : Utilizado para aplicar uma determinada função em umconjunto especifico de dados
# lambda : permite criar funções sem, a necessidade de dar nome as mesmas, funções anonimas
```
